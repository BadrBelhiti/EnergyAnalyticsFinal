```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
from sklearn.model_selection import train_test_split
import zipfile

%matplotlib inline
```

The following method will compile the features used to maximize oil production.

```python
def findInfo(well_name): #This method assumes you have the 'well_production' zip file in the same directory as this notebook
    zf = zipfile.ZipFile(r'well_production.zip')
    data = pd.read_csv(zf.open(well_name+'.csv'))
    
    """
    This method will return the following as a list:
    A. Well length
    B. Number of frac stages
    C. Amount of proppant at each stage
    D. Pump rate
    """
    
    well_length = data['easting'].max() - data['easting'].min() # Difference between max easting and min easting
    
    proppant_points = data.notna()  # This process finds the number of non-NaN rows
    proppant_data = data[proppant_points.all(axis=1)]
    frac_stages = len(proppant_data)
    
    avg_proppant = proppant_data['proppant weight (lbs)'].sum()  # using the previous rows, we can compute the following
    
    avg_pump_rate = proppant_data['pump rate (cubic feet/min)'].sum()
    
    return well_length, frac_stages, avg_proppant, avg_pump_rate
```

Simple example for finding information on the graph:

```python
findInfo('Tarragon 4-119H')
```

Read in the production data for predictions:

```python
data = pd.read_csv(r'../../data/wells/well production.csv')
data.head(10)
```

Finds and stores feature values on individual data sets:

```python
n = len(data['well name'])

for i in range(n):
    i = int(i)
    length, stages, proppant, pump_rate = findInfo(data['well name'][i])
    data.at[i, 'well length (ft)'] = length
    data.at[i, 'frac stages'] = stages
    data.at[i, 'total proppant (lb)'] = proppant
    data.at[i, 'total pump rate (cubic feet/min)'] = pump_rate

data.head(10)  # prints out what the data now looks like
```

Now we will plot each feature against production to find any visual trends for use later.

```python
def plot_feature(feature):
    plt.xlabel(feature)
    plt.ylabel('oil 1')
    plt.title(feature)
    plt.scatter(data[feature], data['oil 12'])
```

```python
plot_feature('well length (ft)')
```

```python
plot_feature('frac stages')
```

```python
plot_feature('total proppant (lb)')
```

```python
plot_feature('total pump rate (cubic feet/min)')
```
