# Notebook to combine all well data for visualization

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import glob
import os

%matplotlib inline
```

Read in well production data and add column with cumulative first year production

```python
wells = pd.read_csv('../../data/wells/well production.csv')

cumulative_productions = []

for index, row in wells.iterrows():
    cumulative_productions.append(row.values[3:15].sum())

wells['cumulative production'] = cumulative_productions
wells.head()
```

Add well's cumulative production and name to all_wells spreadsheet

```python
dataframes = []
directory = '../../data/wells/'

for index, row in wells.iterrows():
    well_name = row['well name']
    well = pd.read_csv(directory + well_name + '.csv')
    well['name'] = well_name
    well['cumulative production'] = row['cumulative production']
    dataframes.append(well)
        
data = pd.concat(dataframes)
data.to_csv('../../data/all_wells.csv', index=False)
```

```python
data.head()
```

```python
prospective_sites = pd.read_csv(r'../../data/prospective_wells.csv')
prospective_sites = prospective_sites.head(100)
prospective_sites
```

Plot a certain feature with the option of plotting any prospective wells, using alpha value to indicate production level

```python
def plot_feature(feature, show_prospective_sites=False):
    plt.xlabel('easting')
    plt.ylabel('northing')
    plt.title(feature)
    plt.scatter(data['easting'], data['northing'], c=data[feature])
    
    if show_prospective_sites:
        alpha = 1
        for index, row in prospective_sites.iterrows():
            plt.scatter(row['easting'], row['northing'], c='r', alpha=alpha)
            alpha -= 0.01
    
    plt.colorbar(label=feature)
```

Plot all prospective wells specified by index in 'well_nums'. Last well placed is colored red to visually assist the person plotting the prospective wells.

```python
def plot_prospective(well_nums):
    plt.xlabel('easting')
    plt.ylabel('northing')
    plt.title('Wells')
    plt.scatter(data['easting'], data['northing'], c='b')
    
    production = 0
    for well_num in well_nums:
        row = prospective_sites.iloc[[well_num]]
        plt.scatter(row['easting'], row['northing'], c='r' if well_num == well_nums[-1] else 'y')
        production += row['production'].values[0]
        
    print(production)
```

Plot first 'n' prospective wells, using alpha to indicate production level.

```python
def plot_wells(n=10):
    plt.xlabel('easting')
    plt.ylabel('northing')
    plt.title('Wells')
    plt.scatter(data['easting'], data['northing'], c='b')
    
    alpha = 1
    for index, row in prospective_sites.iterrows():
        plt.scatter(row['easting'], row['northing'], c='r', alpha=alpha)
        alpha -= 0.01
            
        if index + 1 == n:
            break
```

Plot existing wells and prospective wells

```python
plot_wells(3)
```

The indices of the 10 selected wells, plotted along the previous 100 wells, and aggregate first year cumulative production printed.

```python
prospective = [1, 3, 5, 6, 10, 11, 13, 15, 17, 18]
plot_prospective(prospective)
```
